# AP CSA FRQ Breakdown: Hailstone Sequence

---

## 1. The Official Question

You will implement 3 static methods in a `Hailstone` class that deal with a mathematical sequence called the **Hailstone sequence**.

The sequence is defined as:
- If `n == 1`, the sequence ends.
- If `n` is even â†’ next = `n / 2`
- If `n` is odd  â†’ next = `3n + 1`



### Part (a) â€” `hailstoneLength(int n)`
Return the **number of terms** in the hailstone sequence starting at `n`.  
Ex: `hailstoneLength(5)` â†’ `6` because the sequence is `5, 16, 8, 4, 2, 1`.



### Part (b) â€” `isLongSeq(int n)`
Return `true` if the sequence is **long**, meaning its length is **greater than the starting value**.  
Ex: `isLongSeq(5)` â†’ `true` because `6 > 5`.  
    `isLongSeq(8)` â†’ `false` because `4 â‰¤ 8`.



### Part (c) â€” `propLong(int n)`
Return the **proportion of long sequences** from `1` through `n`.  
Use `isLongSeq` in your implementation.  
Ex: `propLong(10)` â†’ `0.5` because 5 of the first 10 numbers produce long sequences.

## ðŸ” 2. How to Approach This (Based on Top Student and Teacher Tips)

These are best practices based on Reddit, AP teachers, and Runestone guides:

-  **Use loops over recursion** to avoid stack overflow and make it easier to count steps.
-  In part (a), **loop until `n == 1`**, updating `n` based on parity.
-  In part (b), call `hailstoneLength(n)` and compare the result to `n`.
-  In part (c), use a counter and loop from 1 to `n`, checking `isLongSeq(i)`. Return `count / (double) n` for the proportion.
-  **Avoid hardcoding values**. Your methods should work for any `n > 0`.

##  4. Complete Code

```java
public class Hailstone {

    /** Returns the length of a hailstone sequence starting at n */
    public static int hailstoneLength(int n) {
        int count = 1;
        while (n != 1) {
            if (n % 2 == 0) {
                n = n / 2;
            } else {
                n = 3 * n + 1;
            }
            count++;
        }
        return count;
    }

    /** Returns true if the hailstone sequence length is greater than the starting value */
    public static boolean isLongSeq(int n) {
        return hailstoneLength(n) > n;
    }

    /** Returns the proportion of the first n sequences that are long */
    public static double propLong(int n) {
        int longCount = 0;
        for (int i = 1; i <= n; i++) {
            if (isLongSeq(i)) {
                longCount++;
            }
        }
        return longCount / (double) n;
    }
}
