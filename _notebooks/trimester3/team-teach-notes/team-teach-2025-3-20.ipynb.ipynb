# Beneficial and Harmful Effects of Computing
## Benefits
### 1. Medical Advancements
- Improved medical procedures
    - Information is shared to more people and more procedures will become known.
- Data-driven healthcare 
    - Datasets can be shared and a better understanding of illnesses, symptoms, and cures will be found

### 2. Enhanced Business Efficiency
- Data analytics
    - Amount of employees, customers, hours worked, profits made, are all examples of data that can be analyzed
- Automation, logging information
    - Processes like packaging, manufacturing, and receiving orders can be automated through computing.

### 3. Artistic and Creative Expansion
- Digital art platforms 
    - Artists can save their art online and share them to other artists.
- Creative collaboration
    - Artists can share their artworks and work on them simutaneously using different apps and platforms.

### 4. Convenience and Efficiency
- Fast Delivery Services
    - Using drones
- Aerial Photography
    - Using Drones and mounted cameras on cars for example

## Self Driving Cars
### Beneficial Effects
- Increased Safety
    - reduces human error
    - advanced sensores with help of AI can respond faster to dangers
- Traffic Efficiency
    - communication between cars that optimizes traffic
    - reduction in congestion as cars will have similar speed and spacing
- Accessibility
    - People who are elderly or disabled can use these cars
- Environmental Benefits
    - driving optimizing routes which reduces emissions

### Harmful Effects
- Jobs Change
    - ubers, lyfts will decrease, but there can be 
- Security and Privacy Risks
    - Vulnerable to hacking
    - Increased data collection raises concerns abobut privacy
- Ethical and Legal Challenges
    - Accidents can happen
    - AI can be programmed to take certain decisions that may raise ethical concerns

### Unintended Consequences
- Dopamine Feedback Loops from social media
- Prolonged Dopamine-driven activities leads to depression

### Prevention
- limit screen time, use technology intentionally
- Be mindful and self-aware of tech use
