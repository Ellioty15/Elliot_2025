# **Homework Hack #1: Rethinking AI for New Uses**  

## **Task:**  
- AI is often used in new ways that weren’t planned. Your job is to think of a new way to use an AI invention.  

## **My Answer:**  
### **Chatbots**  
- **Original Use:** The chatbot would reply to user input in manners similar to humans.  
- **New Use:** Chatbots can be used differently to moderate conversations.  
  - **Example:** An AI in a Discord server would moderate conversations, ensuring that they do not break the server rules.  
- **Benefits:**  
  - This chatbot would remove the need for human moderators to check conversations manually.  
  - It would ensure a safer space for users, especially if it can react and interact with other messages to delete inappropriate content.  
- **Risks:**  
  - If hacked, this chatbot could ruin the server.  
  - It could also respond to messages unnecessarily and start deleting random messages without context.

# **Homework Hack #2: Ethical AI Coding Challenge**  

## **Task:**  
- AI can cause problems if not designed well. Your job is to find a problem AI has caused and think of ways to fix it.  

## **My Answer:**  
### **Problem: Overdepending on AI**  
- Being able to access AI so easily means that people may use AI all the time to do their work for themselves.  
- People may lose track of how to do work independently or forget how to think for themselves.  
- When AI can't be used, those who are overdependent will struggle to function together.  

### **Two policies that can reduce these risks:**  
1. Have AI reply only to **specific questions**, ensuring that users have prior knowledge and have done research before asking.  
2. Prevent AI from providing **full answers to homework assignments**, as many students use AI to complete their work instead of learning from it.  

## **Write a Short Reflection (3-5 sentences) – Why is ethical AI development important?**  
Ethical AI development is important because the way that AI is developed will determine how the world works. As AI technology advances, both the benefits and risks should be considered. No matter how large the benefits may be, if they are temporary or if there is a risk that will lead to a decline in human health (physical or mental) or impacts the environment in some way, then it is important that the AI be revised and further researched to ensure benefits remain safe for users.

# **Homework Hack #3: AI & Unintended Consequences Research Task**  

## **Task:**  
- Many AI systems have unexpected effects on society. Your goal is to find and analyze a real-world AI example that led to unintended consequences.  

### **[AI-related news story researched](https://www.corralldesign.com/writing/ai-harm-hypocrisy){:target="_blank"}**  

## **Summary:**  
- AI can be used to create art by simply inputting a prompt. This has caused problems such as:  
  - **Copyright Issues:** Artists' works are being duplicated in similar manners by AI, leading to legal concerns.  
  - **Creativity Decline:** Art is becoming less creative, and people are being encouraged to think similarly to how computers think.  
- **Possible Solution:**  
  - Copyright issues can be hard to prevent as AI-generated art can resemble existing works. However, one way to avoid this is to **restrict AI from referencing existing works** when generating new pieces.  
- **Ethical Dilemma:**  
  - The issue of inputting a prompt to generate an illustration is difficult for governments or companies to regulate.  
  - The art community faces an ongoing ethical debate about whether AI-generated work should be considered "art."  

### **Problems that were not mentioned in this article:**  
- AI-generated images of real people can be manipulated to make it seem like they did things they never actually did.  
- This can cause insecurity, leading to **depression, anxiety, or even suicide** for those affected.  

## **Preventing These Issues:**  
- Companies should **prevent AI from generating images that replicate real people or existing artwork.**  
- AI art services should **collaborate with users and artists** to ensure they meet ethical and industry standards.
